```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Dense, Flatten, Conv2D, MaxPooling2D
from sklearn.model_selection import KFold, StratifiedKFold
import pickle
from random import seed
import numpy as np

X_train = pickle.load(open('datasets/X_train.pickle', 'rb'))
y_train = pickle.load(open('datasets/y_train.pickle', 'rb'))

y_train_id = np.array([attrs[0] for attrs in y_train]).reshape(-1,)
y_train_eth = np.array([attrs[1] for attrs in y_train]).reshape(-1,)
y_train_gend = np.array([attrs[2] for attrs in y_train]).reshape(-1,)
y_train_side = np.array([attrs[3] for attrs in y_train]).reshape(-1,)

X_test = pickle.load(open('datasets/X_test.pickle', 'rb'))
y_test = pickle.load(open('datasets/y_test.pickle', 'rb'))

y_test_id = np.array([attrs[0] for attrs in y_test]).reshape(-1,)
y_test_eth = np.array([attrs[1] for attrs in y_test]).reshape(-1,)
y_test_gend = np.array([attrs[2] for attrs in y_test]).reshape(-1,)
y_test_side = np.array([attrs[3] for attrs in y_test]).reshape(-1,)

X_train = X_train/255.0
X_test = X_test/255.0

X_shape = X_train.shape[1:]
X_shape
```

```python
REPEATS = 20

def buildCnnModelGend(outputs, neurons):
    
    cnnModelGend = Sequential([
        #cnn
        Conv2D(filters=32, activation="relu", kernel_size=(3,3), input_shape=X_shape),
        MaxPooling2D((2,2)),

        Conv2D(filters=32, activation="relu", kernel_size=(3,3)),
        MaxPooling2D((2,2)),

        #dense
        Flatten(),
        Dense(neurons, activation="relu"),
        Dense(neurons, activation="relu"),
        Dense(outputs, activation="softmax")
    ])

    cnnModelGend.compile(optimizer="adam",
                 loss="sparse_categorical_crossentropy",
                 metrics=["accuracy"])
    
    return cnnModelGend


def trainModel(numberOfOutputs, numberOfEpochs, y_train, y_test, neurons):
    scores = []
    for i in range(REPEATS):
        print(i)
        tf.keras.backend.clear_session()
        cnnModelGend = buildCnnModelGend(numberOfOutputs, neurons)
        cnnModelGend.fit(X_train, y_train, epochs=numberOfEpochs)
        score, acc = cnnModelGend.evaluate(X_test, y_test)
        scores.append(round(acc,4))
        print(round(acc,4))
    return scores

def printScores(scores):
    print(f'{scores} \n Mean: {np.mean(scores)}\n Std. {np.std(scores)}')
```

```python
# ID model

scores = trainModel(100, 15, y_train_id, y_test_id, 256)
```

```python
printScores(scores)
```

```python
# ETHNICITY model

scores = trainModel(6, 15, y_train_eth, y_test_eth, 128)
```

```python
printScores(scores)
```

```python
# GENDER model

scores = trainModel(2, 10, y_train_gend, y_test_gend, 128)
```

```python
printScores(scores)
```

```python
# LEFT RIGHT model - prej je blo 15 epoch, 128 neurons

scores = trainModel(2, 10, y_train_side, y_test_side, 512)
```

```python
printScores(scores)
```
